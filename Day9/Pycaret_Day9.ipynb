```python
df=pd.read_csv('PastHires.csv')
```

```python
df
```

```python
# pycaret.nlp import *   pycaret ın nlp paketini import etme
#                        lda=Latent Drichlet Allocation - Topic Modelling - Hangi Konuda Konuşulduğunu Anlama
#                        NER - Named Entity Recognition - Cümle içerisinde özne yüklem fiilleri anlayarak Organizasyon Firma isimlerini çıkartmaktadır
#                        Markoc Chain - önceki yazışmalara bakarak yeni konuşma ya da yazıların tahmin edilmesi
```

```python
df["Employed?"]=df["Employed?"].map({"Y":1,"N":0})
```

```python
df
```

```python
df=df.replace({"Y":1,"N":0,"BS":1,"MS":2,"PhD":3})
```

```python
df
```

```python
df.corr()
```

```python
y=df["Hired"] #İşe alınıp alınmaması durumu 
```

```python
x=df.drop("Hired",axis=1)
```

```python
x
```

```python
from sklearn.tree import DecisionTreeClassifier
d=DecisionTreeClassifier()
```

```python
d.fit(x,y)
```

```python
d.predict([[10,1,4,1,0,1]]) #özellikleri bu olan kişiyi işe alınsın mı alınmasın mı? 
```

```python
#Cevap 1 döndü yani alınması uygun firma ile uyumlu çalışabilecek kişi
```

```python
from sklearn.decomposition import PCA  #Veri boyutunu küçültme yani 6 sütunlu bir tablo yerine 2 sütuna indirgeyerek 2 sütun üzerinde çalışıyoruz
```

```python
pca=PCA(n_components=2)
```

```python
x_y=pca.fit_transform(x)
```

```python
x_y
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_) #verilerin %97 si temsil ediliyor %3 lük bir kayıp var
```
