```python
df["zipcode"]=df["zipcode"].astype("category")
```

```python
df=df[df['bedrooms']<10]
```

```python
df=df[df['bathrooms']<7]
```

```python
df['bedrooms']=df['bedrooms']**2
```

```python
df['bathrooms']=df['bathrooms']**2
```

```python
df["age"]=2022-df["yr_built"]
```

```python
#sns.pairplot(df);
```

```python
import numpy as np
df['basement']=np.where(df['sqft_basement']>0,1,0)
df['renovated']=np.where(df['yr_renovated']>0,1,0)
```

```python
x=df[["sqft_living","grade","sqft_above","sqft_living15","bathrooms","view","sqft_basement","bedrooms","lat","waterfront","floors","age","zipcode"]]
y=df["price"] 
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.60,random_state=42)
```

```python
x_train.shape,x_test.shape 
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr=LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
tahmin=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error 
```

```python
(mean_squared_error(tahmin,y_test))**0.5
```

```python
r2_score(tahmin,y_test)
```
