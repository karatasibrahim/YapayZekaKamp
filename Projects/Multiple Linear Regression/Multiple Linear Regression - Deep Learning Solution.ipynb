```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

pd.set_option("display.max_columns",100)
df=pd.read_csv("advertising.csv")
df
```

```python
df.head()
```

```python
df.Sales
```

```python
x= df.drop("Sales",axis=1) #price hariç diğer satırlar
y=df["Sales"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=Sequential()
model.add(Dense(19,activation="relu")) #dense layer tüm nöronlar birbirine bağlanmış demektir.
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(1))
model.compile(optimizer="Adam",loss="mse")
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=500)
model.summary()
```

```python
loss_df=pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
tahmin=model.predict(x)
```

```python
r2_score(tahmin,y)
```

```python
 (mean_squared_error(tahmin,y))**0.5
```
