```python
#Shapiro Test
```

```python
from scipy.stats import shapiro
```

```python
def shapiroTest(ts):
    val=shapiro(ts)
    statstic=round(val[0],3)
    pval=round(val[1],3)
    criticalVal=0.05
    if pval>criticalVal:
        print("a Gaussian case")
    else:
        print("a non Gaussian case")
```

```python
shapiroTest(myPrice)  # Normal dağılmamış  Çan eğirisine benzese de veiler normal dağılmamış
```

```python
#Rolling Mean => Kayan Ortalama
```

```python
def rolling_Mean_STD(timeseries):
    rolmean=timeseries.rolling(window=12).mean()
    rolstd=timeseries.rolling(window=12).std()
    
    orig=plt.plot(timeseries, color="blue",label="Original")
    mean=plt.plot(rolmean, color="red",label="Rolling Mean")
    std=plt.plot(rolstd, color="black",label="Rolling Std")
    plt.legend(loc="best")
    plt.title("Rolling Mean & Standard Deviation")
    plt.show(block=False)
```

```python
rolling_Mean_STD(myPrice)
```

```python
def mean_checjer(timeSeries):
    thres=int(len(timeSeries)/3)
    firstPart=timeSeries[1:thres]
    secondPart=timeSeries[(thres+1):(2*thres)]
    thirdPart=timeSeries[(2*thres+1):]
    
    print("the length of the first part is", len(firstPart))
    print("the length of the first part is", len(secondPart))
    print("the length of the first part is", len(thirdPart))
    
    print(".................................................")
    
    firstMean=firstPart.mean()
    secondMean=secondPart.mean()
    thirdMean=thirdPart.mean()
    
    print("..................................................")
    
    print("the length of the first part is", firstMean)
    print("the length of the first part is", secondMean)
    print("the length of the first part is", thirdMean)
    
    
```

```python
mean_checjer(ts)
```

```python
from statsmodels.tsa.stattools import adfuller
def stationarity_test(timeseries):
    print("Results of Dicket-Fuller Test: ")
    dftest=adfuller(timeseries,autolag="AIC")
    dfoutput=pd.Series(dftest[0:4],index={"Test Statictic","p-value","#lags Used","Number of Observations User"})
    
    for key,value in dftest[4].items():
        dfoutput["Critical Value (%s)"%key]=value
    print(dfoutput)
```

```python
stationarity_test(myPrice)
```

```python
# ACF and PACF
```

```python
def comb_p_d_q(pVals,dVals,qVals):
    return {(p,d,q) for p in pVals for d in dVals for q in qVals}
```

```python
comb_p_d_q([0,1,2],[0,1,2],[0,1,2])
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
def aic_finder(ts,model_order):
    model=ARIMA(ts, order=model_order)
    model_fit=model.fit(disp=0)
    aicval=model_fit.aic
    if np.isfinite(aicval):
        return aicval
    else:
        return 9999
```

```python
aic_finder(myPrice.values,(2,0,1))
```
