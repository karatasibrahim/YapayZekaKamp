```python
df["Date"]=pd.to_datetime(df["Date"]) #Date alanının değerini object ten detatime a çevirdik
```

```python
# Feature Engineering: Verinin aslında olmayan ancak veiden yararlanılarak üretilen sütun işlemine denir
```

```python
df.info()
```

```python
df['day']=(df['Date']).dt.day
df['month']=(df['Date']).dt.month
df['year']=(df['Date']).dt.year
```

```python
df['month_name']=df['Date'].dt.month_name()
df['weekday']=df['Date'].dt.day_name()
```

```python
df['Time']=pd.to_datetime(df['Time'])
df['Hour']=(df['Time']).dt.hour
```

```python
df['Hour']=(df['Time']).dt.hour
```
