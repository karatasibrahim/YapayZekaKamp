```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(34) #aynı rakamlar üzerinde çalıştık
x=np.arange(20) #
y=np.random.normal(10,1,20)
z=np.random.normal(10,2,20)
k=np.random.normal(10,3,20)
```

```python
fig=plt.figure(figsize=(10,15)) # gafiklerin width 10 height 15olarak belirledik 
ax=fig.add_subplot(311) #üç satır 1 sütun
ax.set_title("x in y e göre değişimi")
ax.set_xlabel("X verileri")
ax.set_ylabel("Y verileri")
ax.plot(x,y)

bx=fig.add_subplot(312) 
bx.set_title("x in z ye göre değişimi")

bx.plot(x,z,color='yellow')

cx=fig.add_subplot(313) 
cx.set_title("x in k ye göre değişimi")
cx.plot(x,k)
```

```python
df["Date"]=pd.to_datetime(df["Date"]) #Date alanının değerini object ten detatime a çevirdik
```

```python
# Feature Engineering: Verinin aslında olmayan ancak veiden yararlanılarak üretilen sütun işlemine denir
```

```python
df.info()
```

```python
df['day']=(df['Date']).dt.day
df['month']=(df['Date']).dt.month
df['year']=(df['Date']).dt.year
```

```python
df['month_name']=df['Date'].dt.month_name()
df['weekday']=df['Date'].dt.day_name()
```

```python
df['Time']=pd.to_datetime(df['Time'])
```

```python
df['Hour']=(df['Time']).dt.hour
```
